```python
import CodeBase.DatosAgrupados as da
import CodeBase.Edwin as ed
import CodeBase.GenerarGraficas as gg

# DATOS
datos = [0,0,0.33,0.33,0.33,0.33,0.66,0.66,0.66,0.66,0.99,0.99,0.99,1.32,1.32,1.32,1.65,1.65,1.65,1.65]

# CALCULAR DATOS
clases, limsInf, limsSup, mrksClases, fa, fr, frAc = da.generateGroupedData(datos)
frStr, frAcStr = [], []
for i in range(len(fr)):
    frStr.append(str(fr[i]) + "%")
for i in range(len(frAc)):
    frAcStr.append(str(frAc[i]) + "%")
    
# IMPRIMIR TABLA
ed.printHTMLTable(["Clases", "Límites inferiores", "Límites superiores", "Marca de clase", "F<sub>a</sub>", "F<sub>r</sub>", "F<sub>r</sub> Acum."],[clases, limsInf, limsSup, mrksClases, fa, frStr, frAcStr])

# MOSTRAR GRAFICAS
gg.displayCharts(mrksClases, fa, fr, frAc)
```
